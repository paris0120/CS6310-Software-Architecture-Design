### Experiments

## Sections:
### 1. Context: Background, Motivation, Vocabulary
1. provides background and motivation for the study
1. define any specialized vocabulary necessary

### 2. Research Questions
    1. Examine the tradeoffs between various non-functional requirements
        1. i.e. space and time
        1. each tradeoff can be expressed in the form of a question
            1. how are execution times and memory footprint effected as teh amount of pre-processing computations vary
    1. The second section lists such research questions
        1. each question 
            1. should be formulated in a neutral fashion regarding the dependent variable being measured
            1. indicate what factors are being varied (independent variables)
            1. should be numbered for later reference in your report

### 3. Subject
1. in experiments, a subject is something that you are studying, usually a program
1. a design study compares multiple subjects
1. In the third section, each subject should be briefuly described, defferentiating it from hte other subjects


### 4. Experimental Conditions
1. the fourth section is titled Experimental Conditions
    1. A software design study normally means running several versiosn of a program, making measurements and evaluating the results
    1. These programs' executions take place on computers configured with resources
        1. number of cores
        1. amount of RAM
        1. clock speed
        1. potentially the networking
    1. To support the goal of repeatability, this configuration information should be explicitly documented in the report
    1. describes the environment of which the study will take place
        1. the machiens
        1. their models
        1. operating system
        1. programming languages
        1. any virtual machines and their versions
        1. the network
        1. the build and execution parameters
        1. input files
        1. confounding factors
            1. other users on the machines at the same time 
            1. other processes going on

### 5. Variables
1. the fifth section is titled variables
    1. the independent and dependent variables must be identified and appropriate metrics specified
    1. design study like experiments allow designers to alter conditions and results
    1. the altered conditions comoprise the independent variables
    1. corresponding results comprise the dependent variables 
    1. describes both independent and dependent variables the units of measure and how the resaerch questions address them
    1. each variable has 
        1. a unique name
        1. a description
        1. a unit of measurement such as seconds
    1. include a summary table with three columns, 
        1. for each research question 
        1. the independent variable
        1. dependent variable


        

### 6. Method
1. include
    1. the number of trials
    1. measurement devices
    1. tools
    1. randomization technique
    1. number o significant digits used in the measurements
    1. an explicit statement of which subjects will be run and the arguments used
        1. study the relationship of performance to grid size
            1. what different grid sizes you will be using 
    1. briefly describe any statistical techniques you will use
        1. linear regression
        
### 7. Results
1. the point of conducting a design study is to produce data
1. presents the data collected and their statistical analysis 

### 8. Discussion
1. opportunity to interpret the data you collected and provide a discussion of its implications
    1. offer an explanation for of any unexpected values
1. allows you to reflect on the experimentaion itself
    1. any suggestions
    1. any suggested further work
    1. improving the study process itself
    
### 9. Conclusions
1. Summarize the results and draw conclusions
1. you should provide explicit answers to the research questions you raised in the second section
