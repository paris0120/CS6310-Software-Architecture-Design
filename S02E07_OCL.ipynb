### OCL
1. Object Constraint Language
1. Offical part of UML
1. A specification language 
1. Strongly typed declarative specification of system properties 
1. Consist of a means to express
    1. constraints 
    1. some collection classes
    1. UML diagram navigation
1. Supported by various tools
    1. Rational Rose
    1. ArgoUML
    1. Eclipse
    1. Poseidon/Octopus
    1. Enterprise Architect

### Why Do We Need OCL?
1. UML diagrams are limited in what they can express
    1. Structural relationships
    1. Behavioral descriptions
1. Mechanism for specifying precise semantics
1. OCL extends UML with
    1. Class invariants
    1. Operation pre and post conditions
    1. Guards on state-machine transitions

### OCL Overview
1. Declarative not procedural
    1. pure expression language
        1. describe values
    1. No assignement statement of other side effects
    1. specifies 
        1. assertions 
        1. constraints 
        1. properties
        1. usually with equal signs 
1. Strongly typed
    1. Built-in primitive types and types introduced in UML diagrams
1. has only one key concept: constraint
    1. Highest level mechanism is the constraint
1. Constraint
    1. Formal assertion of system properties 
    

### Uses of OCL
1. To specify invariants on classes in the class model
1. To describe pre- and post conditions on Operations and Methods
1. To specify derivation rules for derived attributes
1. To describe guards on transitions in Statechart diagrams 
1. To specify targets for messages and actions
1. To specify type invariants for Stereotypes
1. As a query language

### Syntax
1. context \<identifier> \<constraintType> \<Boolean expression>
1. Simple 
1. Just one statement
1. identifier
    1. usually a name of a class
    1. sometimes might be the name of a method in a class
1. constraintType
    1. keywords
    1. 3 types of constraint
        1. inv: invariance
            1. Statement of a property that is always true
            1. Express key system requirements
            1. inv keyword            
            1. Examples
                1. context LargeCompany inv: numberOfEpmloyees>50
            1. Role of invariants
                1. Concept of integrity constraints
                    1. dangling pointer
                    1. properties of the data in the database that describe whether the databse is sound
                1. OCL invariant constraints
                    1. play this role for system specifications
        1. pre: precondition
            1. what must be true for the operation to meaningfully take place
            1. example
                1. the argument to the squre root routine must be non-negative
        1. post: postcondition
            1. what is guaranteed to be true after the Operation completes
            1. example
                1. the square of the conputed result must equal the argument
            1. used to describe the results of changes in the values of an attribute
            1. example
                1. context Account::deposit(Real: amount)
                1. pre: amount > 0 
                1. post: balance = balance@pre + amount
1. Boolean express
    1. actual constraint
    
1. SQRT Expressed in OCL
    1. context Real::squareRoot(): Real
    1. pre: self>=0
    1. post: self = result * result

### Let clause
1. Way of introducing an abbreviation into a constraint
1. Has two parts
    1. Both expressions 
1. example:
    1. let income: Interger = self.job.salary->sum() in 
        1. if isUnemployed then income<100
            1. else income >=100
        1. endif  

### Navigatin
1. OCL constraints are assoicated with class model diagrams
1. Each constraint specifies a context
1. To refer to features of a non-context class in an OCL invariant, you must use a qualified name
1. Navigation is done by walking through a diagram from the context class to the other class using intermediary relationship lines and class rectangles 

#### Navigation example
1. assume customer is the context class

1. order.data: self.order.date

#### Navigation: Multiplicity
1. WHen an association has multiplicity greater than one the result is a collection rather than an individual object
1. If an Operation is performed, thenthe -> symbol is used to seperate the collection name from the Operation name

## Collections

    1. collection (primary class)
        1. provides features that hold true of all collections
            1. size
            1. includes
            1. count
            1. includesAll
            1. isEmpty
            1. notEmpty
            1. sum
            1. exists
            1. forAll
            1. iterate
    1. sets
    1. bags
    1. sequences
    1. provide specialized Operations for set, bag and sequence
